# Combining Dataframes to form nescessary metrics for selection of Best 11

```python
# DataFrame for attack
df_attack = pd.merge(df,df_shoot,left_on=["player","team","position"],right_on=["player","team","position"],how="inner").merge(df_pass,on=["player","team","position"]).merge(df_gca,on=["player","team","position"]).merge(df_pos,on=["player","team","position"]).merge(df_misc,on=["player","team","position"])
```

```python
# Putting zero in place of NaN
df_attack.fillna(0,inplace=True)
df_attack.head(10)
```

```python
# DataFrame for midfield and defense
df_mid_def=pd.merge(df_attack,df_def,left_on=["player","team","position"],right_on=["player","team","position"],how="inner")
```

```python
# Putting zero in place of NaN
df_attack.fillna(0,inplace=True)
df_attack.head(10)
```

```python
# DataFrame for goakeeper
df_keeper=pd.merge(df_gk,df_gkadv,left_on=["player","team","position"],right_on=["player","team","position"],how="inner")
```
